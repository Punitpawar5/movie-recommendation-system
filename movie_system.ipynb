```python
def convert_cast(obj):
  l=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      l.append(i['name'])
      counter +=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def find_director(obj):
  l=[]
  counter = 0
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(find_director)
```

```python
# each tag will convert to VECTOR


# use BAG OF Words
"""
large text = tag1+tag2.... >> get 5000 words similar
in first movie check how many time 'Action' come and same other Words then other MOVIES
   w1 +w2+ w3.... w5000
m1 5    3   4      0
m2 4    1   4      5

so this m1 is single vector we will get the top distance like m1 has closest m3,m45, m5...


m1 5    3   4      0
# not consider stop word like: and, of, then, a, is, etc....

"""
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors= cv.fit_transform(df['tag']).toarray()
```

### Steming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps =PorterStemmer()
```

```python
ps.stem("loved")
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
df['tag']=df['tag'].apply(stem)
```

###Got Stem

```python
vectors= cv.fit_transform(df['tag']).toarray()
cv.get_feature_names_out()[:80]
```

```python
df.shape
```

4806 movies and same vectors in 5k dimension

we should not use Eucleadian Distan in High Dimension
we should use Consine Distance
dis inversly similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# cosine_similarity(vectors).shape 1 movie with 4806 same  (4806 movies with , 4806)
similarity = cosine_similarity(vectors)
```

```python
similarity[0] #  0 index score with every 4806 movies
```

```python
# sorted(similarity[0], reverse=True)[1:5] not use because our index is changes of movie to get it so we use ENUMERATE

sorted(list(enumerate(similarity[0])), reverse=True, key= lambda x:x[1])[1:6]
```

```python
df[df['title']=='Avatar'].index[0]
```

```python
df.iloc[1216].title
```

```python
def recommend(movie):
  move_index = df[df['title']==movie].index[0]
  distance = similarity[move_index]
  movie_list= sorted(list(enumerate(distance)), reverse=True, key= lambda x:x[1])[1:6]
  for i in movie_list:
    print(df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
